```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8, 6))
plt.hist(df['diamond_clarity'], bins=20)
plt.xlabel('diamond_clarity')
plt.ylabel('frequency')
plt.title('Counts of Diamond')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(df['diamond_weight'], bins=20)
plt.xlabel('diamond_weight')
plt.ylabel('frequency')
plt.title('Distribution of Diamond Weight')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
df['diamond_color'].value_counts().plot(kind='bar')
plt.xlabel('Country')
plt.ylabel('Count')
plt.title('No. of Country')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['diamond_weight'], df['price'])
plt.xlabel('diamond_weight')
plt.ylabel('price')
plt.title('Diamond Wieight VS Price')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(df['diamond_weight'], bins=20)
plt.xlabel('diamond weight')
plt.ylabel('frequency')
plt.title('Fq. of Daimond Weight')
plt.show()
```

```python
from pandas.plotting import autocorrelation_plot
plt.figure(figsize=(8, 6))
df['diamond_color'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Pie of Diamond Colour')
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
df_grouped = df.groupby(['diamond_shape', 'diamond_cut_grade']).size().unstack()
df_grouped.plot(kind='bar', stacked=True)
plt.xlabel('diamond hsape')
plt.ylabel('count')
plt.title('Comparision os Diamond Shape & Cut Grade')
plt.legend(title='Diamond Cut Grade')
plt.show()
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='diamond_cut_grade', y='price', data=df)
plt.xlabel('Diamond Cut Grade')
plt.ylabel('Price')
plt.title('Comparision of Diamond Cut Grade')
plt.show()
```

```python
numerical_cols = ['diamond_weight', 'price']
corr_matrix = df[numerical_cols].corr()

plt.figure(figsize=(7, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
numerical_cols = ['diamond_weight', 'price']
sns.pairplot(df[numerical_cols])
plt.title('Pair Plot of Numericals Column')
plt.show()
```
