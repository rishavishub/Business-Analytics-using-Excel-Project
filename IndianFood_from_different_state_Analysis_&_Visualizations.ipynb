#Performing Visualisations

```python
import matplotlib.pyplot as plt
```

```python
food
```

# Types of Food Categories (Histogram)

```python
plt.figure(figsize=(8,6))
plt.hist(food['diet'], bins=20)
plt.xlabel('Food category')
plt.ylabel('Counts')
plt.title('Types of Food Categories')
plt.show()
```

# Maximum Prep_time Taken (Histogram)

```python
plt.figure(figsize=(8,6))
plt.hist(food['prep_time'], bins=20)
plt.xlabel('Time')
plt.ylabel('Duration')
plt.title('Preparation time')
plt.show()
```

# Flavour profile on Course

```python
plt.figure(figsize=(8,6))
plt.scatter(food['flavor_profile'], food['course'])
plt.xlabel('Flavour')
plt.ylabel('Course')
plt.title('Showing Flavour Profile on Course')
plt.show()
```

# Region and There Diet (Pie Chart)

```python
plt.figure(figsize=(8,6))
food['region'].value_counts().plot(kind='pie', autopct='1%.1f%%')
plt.axis('equal')
plt.title('Region and There Diet')
plt.show()
```

# Region on the Basis of Region (Bar)

```python
plt.figure(figsize=(8,6))
food_grouped = food.groupby(['region','flavor_profile']).size().unstack()
food_grouped.plot(kind="bar", stacked=True)
plt.xlabel('Region')
plt.ylabel('Profile')
plt.legend(title='Food Profile')
plt.show()
```

# Comparision betweenn Prep_Time and Cook Time

```python
import seaborn as sns
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='prep_time', y='cook_time', data=food)
plt.xlabel('Cooking Duration')
plt.ylabel('Prepration Duration')
plt.title('Comparision between Preptime and Cooktime')
plt.show()
```

#   Correlation Between Preparation and Cooking (HeatMap)

```python
numerical_cols = ['prep_time', 'cook_time']
corr_matrix = food[numerical_cols].corr()

plt.figure(figsize=(8,9))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

# Pair Plot of Preptime and Cooktime

```python
numerical_cols = ['prep_time', 'cook_time']
sns.pairplot(food[numerical_cols])
plt.title('Pair Plot')
plt.show()
```
